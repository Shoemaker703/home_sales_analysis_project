```python

from sklearn.model_selection import train_test_split
import pandas as pd

df = pd.read_csv('../../data/kc_house_data.csv')
## DROP PRICE, ID!
```

```python
df.head()
```

```python
x = df.drop(columns = ['id', 'price'])
y = df['price']
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y,  test_size=0.3, random_state=42)
```

```python
x_test.to_csv('../../data/kc_house_data_xtest.csv')
x_train.to_csv('../../data/kc_house_data_xtrain.csv')
y_test.to_csv('../../data/kc_house_data_ytest.csv')
y_train.to_csv('../../data/kc_house_data_ytrain.csv')
```
