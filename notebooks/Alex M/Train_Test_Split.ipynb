```python

from sklearn.model_selection import train_test_split
import pandas as pd

df = pd.read_csv('../../data/kc_housing_CLEAN.csv', index_col = 0)
## DROP PRICE, ID!
```

```python
df.info()
```

```python
x = df.drop(columns = ['id', 'price'])
y = df['price']

cat_cols= ['grade', 'condition']
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y,  test_size=0.3, random_state=42)
```

```python
x_train.info()
```

```python
x_train[cat_cols]
```

```python
x_test.describe()
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler 
ohe = OneHotEncoder(sparse = False, handle_unknown = 'ignore') 
print(cat_cols)
x_train_ohe = ohe.fit_transform(x_train[cat_cols])

```

```python
x_train_ohe
```

```python
x_train_ohe_data = pd.DataFrame(x_train_ohe, columns= ohe.get_feature_names(cat_cols), index = x_train.index)

x_train_full = pd.concat([x_train.drop(cat_cols, axis=1), x_train_ohe_data], axis = 1)
x_train_full
```

```python
x_test_ohe = ohe.transform(x_test[cat_cols])
x_test_ohe_data = pd.DataFrame(x_test_ohe, columns= ohe.get_feature_names(cat_cols), index = x_test.index)

x_test_full = pd.concat([x_test.drop(cat_cols, axis=1), x_test_ohe_data], axis = 1)
x_test_full
```

```python
x_test_full.to_csv('../../data/kc_house_data_xtest_FINAL2.csv', )
x_train_full.to_csv('../../data/kc_house_data_xtrain_FINAL2.csv')
y_test.to_csv('../../data/kc_house_data_ytest_FINAL2.csv')
y_train.to_csv('../../data/kc_house_data_ytrain_FINAL2.csv')

## when load in Index col = 0
```
